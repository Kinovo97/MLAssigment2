üéØ What is PCA?
PCA (Principal Component Analysis) is a dimensionality reduction technique used in machine learning and data analysis. It helps:

Reduce the number of features (columns) while

Preserving as much variance (information) as possible.

üß† The Main Idea
PCA finds new axes (called principal components) that:

Are linear combinations of the original features

Are uncorrelated

Capture the maximum variance in the data

üìà What It Does in Practice
Imagine you have 100 features. Many of them may be redundant or correlated. PCA:

Projects the data to a new space (rotated axes).

Keeps only the top k components that explain most of the data's variation.

## Part 3 - Experiments
You could add as many code cells as needed

```python
# Feature engineering
X_test = test_df.drop(columns=['target'])
Y_test = test_df['target']

Y_train = train_df['target']
X_train = train_df.drop(columns=['target'])
```

```python
# Apply MinMaxScaler and StandardScaler
minmax_scaler = MinMaxScaler(feature_range=(-1, 1))
standard_scaler = StandardScaler()

X_train_standard_scaled = standard_scaler.fit_transform(X_train) #StandardScaler
X_test_standard_scaled = standard_scaler.transform(X_test) #StandardScaler
X_train_minmax_scaled = minmax_scaler.fit_transform(X_train) #MinMaxScaler
X_test_minmax_scaled = minmax_scaler.transform(X_test) #MinMaxScaler

KNN = KNeighborsClassifier() #Training
tree_decision = DecisionTreeClassifier()

# Definition of the parameter grids for tree_decision and KNN
knn_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'metric': ['euclidean', 'manhattan']
}

dt_grid = {
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10, 20]
}
```

```python
#Performing the Grid Search
def perform_gridsearch(model, param_grid, X, Y):
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='f1_macro', n_jobs=-1)
    grid_search.fit(X, Y)
    return grid_search.best_params_, grid_search.best_score_


# Original Data
best_knn_params_original, best_knn_score_original = perform_gridsearch(KNeighborsClassifier(), knn_grid, X_train, Y_train)
best_dt_params_original, best_dt_score_original = perform_gridsearch(DecisionTreeClassifier(), dt_grid, X_train, Y_train)

# Standard Scaled Data
best_knn_params_standard, best_knn_score_standard = perform_gridsearch(KNeighborsClassifier(), knn_grid, X_train_standard_scaled, Y_train)
best_dt_params_standard, best_dt_score_standard = perform_gridsearch(DecisionTreeClassifier(), dt_grid, X_train_standard_scaled, Y_train)

# Min-Max Scaled Data
best_knn_params_minmax, best_knn_score_minmax = perform_gridsearch(KNeighborsClassifier(), knn_grid, X_train_minmax_scaled, Y_train)
best_dt_params_minmax, best_dt_score_minmax = perform_gridsearch(DecisionTreeClassifier(), dt_grid, X_train_minmax_scaled, Y_train)

#Creating and displaying summary table
summary_table = pd.DataFrame({
    'Model': ['KNN (Original)', 'Decision Tree (Original)', 'KNN (Standard Scaled)', 'Decision Tree (Standard Scaled)', 'KNN (Min-Max Scaled)', 'Decision Tree (Min-Max Scaled)'],
    'Best Parameters': [best_knn_params_original, best_dt_params_original, best_knn_params_standard, best_dt_params_standard, best_knn_params_minmax, best_dt_params_minmax],
    'Best F1 Macro Score': [best_knn_score_original, best_dt_score_original, best_knn_score_standard, best_dt_score_standard, best_knn_score_minmax, best_dt_score_minmax]
})

summary_table
```

## Part 4 - Training 
Use the best combination of feature engineering, model (algorithm and hyperparameters) from the experiment part (part 3)

```python
best_index = summary_table['Best F1 Macro Score'].idxmax()
best_model_info = summary_table.loc[best_index]

best_model_name = best_model_info.at['Model']
best_params = best_model_info.at['Best Parameters']
best_f1_score = best_model_info.at['Best F1 Macro Score']

print(f"Best Model: {best_model_name}")
print(f"Best Parameters: {best_params}")
print(f"Best F1 Macro Score: {best_f1_score}")
```

```python
if 'Standard Scaled' not in best_model_name:
    X_train_final = X_train_minmax_scaled
    X_test_final = X_test_minmax_scaled
else:
    X_train_final = X_train_standard_scaled
    X_test_final = X_test_standard_scaled
if 'KNN' not in best_model_name:
    best_model = DecisionTreeClassifier(**best_params)
else:
    best_model = KNeighborsClassifier(**best_params)
best_model.fit(X_train_final, Y_train)
```

```python
print("First 5 predictions on test set:", best_model.predict(X_test_final)[:5])
```

```python
pca = PCA(n_components=5)
X_train_pca = pca.fit_transform(X_train_standard_scaled)
X_test_pca = pca.transform(X_test_standard_scaled)

# Grid Search for KNN on PCA-transformed data
grid_pca_knn = GridSearchCV(KNN, knn_grid, scoring='f1_macro', cv=5)
grid_pca_knn.fit(X_train_pca, Y_train)

print("Best Params (PCA-KNN):", grid_pca_knn.best_params_)
print("Best F1 Macro Score (CV):", grid_pca_knn.best_score_)

# Predict on test set
best_knn_pca = grid_pca_knn.best_estimator_
y_pred_pca = best_knn_pca.predict(X_test_pca)
f1_test_score = f1_score(Y_test, y_pred_pca, average='macro')
print("Test F1 Score (PCA-KNN):", f1_test_score)
```

## Part 5 - Apply on test and show model performance estimation

```python
# Evaluate the model on the test dataset
y_pred = best_model.predict(X_test_final)
report = classification_report(Y_test, y_pred)
f1 = f1_score(Y_test, y_pred, average='macro')

print("Classification Report for Best Model:")
print(report)
print(f"Macro F1 Score for Best Model: {f1}")
```
